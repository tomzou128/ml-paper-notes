## Goal
You have a lot of sprite images. You want even more sprites. You need a neural network that can generates more sprites for you.

You want a neural network to learn what a sprite it, so you should provide:
- Fine details
- General outlines
- Eveything in between

One method is to add different noise levels to the training data of sprites (noising process inspired by diffusion in physics)

**What should the neural network be thinking at each level of noise?**

Given an image of a sprite named Bob with different level of noise,

- If it looks very likely to be Bob the sprite, keep Bob as is, it is a sprite.
- If it looks likely to be Bob the sprite, suggest possible details to make the image to just look like Bob the sprite.
- If it only looks like an outline of a sprite, suggest general details for likely sprites, or suggest based on some text prompt.
- If it looks like absolutely random noise, suggest an outline of sprite.


## Train a neural network to make sprites
The neural network learns to take different noisy images and turn them back into sprites. It learns to remove the noise you added. The inital noisy image is sampled from normal distribution.

So when you ask the neural network for a new sprites:
- Sample noise from the normal distribution.
- Get a completely new sprite by using the net to remove the noise.

## Inference
At each step during inference, NN predicts the noise on the noisy image. Then we subtract the predicted noise from the noisy image to get something more sprite like. So realistically, that is just a prediction of part of the noise, and it can't remove all the noise in one step. So we need to remove noise recursively.
